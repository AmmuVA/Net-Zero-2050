## **Import necessary libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## **Carbon emission**

```python
emission=pd.read_excel(r'data.xlsx')
emission
```

## X and Y values

```python
X=emission['Year']
y=emission['Carbon emission']
```

```python
X=X.values.reshape(-1, 1)
```

## Test Train Split

```python
#importing train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

## Import models

```python
#importing the algorithm and fitting the dataset with the model
from sklearn.ensemble import RandomForestRegressor
model =RandomForestRegressor()
model.fit(X_train,y_train)
Y_pred = model.predict(X_test)#Predict testing set
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
import math
from sklearn.metrics import mean_squared_error
dataset=emission
mae=[]
mse=[]
rmse=[]
r2=[]
for n in range(9):
  poly_reg = PolynomialFeatures(degree = n)
  X_poly = poly_reg.fit_transform(X)
  lin_reg_2 = LinearRegression()
  lin_reg_2.fit(X_poly, y)
  y_predicted_poly = lin_reg_2.predict(poly_reg.fit_transform(X))
  dataset['y_predicted_poly']= y_predicted_poly
  mse_poly = mean_squared_error(dataset['Carbon emission'],dataset['y_predicted_poly'])
  print('Degree:',n)
  print('mae:',mean_absolute_error(dataset['Carbon emission'],dataset['y_predicted_poly']))
  mae.append(mean_absolute_error(dataset['Carbon emission'],dataset['y_predicted_poly']))
  print('mse:',mse_poly)
  mse.append(mse_poly)
  print('rmse:',math.sqrt(mean_squared_error(dataset['Carbon emission'],dataset['y_predicted_poly'])))
  rmse.append(math.sqrt(mean_squared_error(dataset['Carbon emission'],dataset['y_predicted_poly'])))
  print('r2_score:',r2_score(dataset['Carbon emission'],dataset['y_predicted_poly']),'\n')
  r2.append(r2_score(dataset['Carbon emission'],dataset['y_predicted_poly']))
  plt.scatter(X, y, color = 'red')
  plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
  plt.title('(Polynomial Regression)')
  plt.xlabel('Position level')
  plt.ylabel('emission rate')
  plt.show()
```

```python
#built the neural network architecture
import tensorflow as tf
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(15, input_dim=15, activation='relu'))
model.add(tf.keras.layers.Dense(15, input_dim=15, activation='sigmoid'))
model.add(tf.keras.layers.Dense(15, input_dim=15, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='linear'))
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['mae'])
```
